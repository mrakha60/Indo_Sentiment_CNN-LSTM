```python
import pandas as pd
from transformers import pipeline

# Baca dataset dari file CSV
file_path = "/content/IKD_Dataset_Processed.csv"
df = pd.read_csv(file_path)

# Muat model pre-trained IndoBERT untuk analisis sentimen
model_name = "ayameRushia/bert-base-indonesian-1.5G-sentiment-analysis-smsa"
classifier = pipeline("text-classification", model="ayameRushia/bert-base-indonesian-1.5G-sentiment-analysis-smsa")

# Pastikan data dalam kolom 'review_text' bertipe string
df['review_text'] = df['review_text'].astype(str)

# Lakukan analisis sentimen pada setiap teks
results = classifier(df['review_text'].tolist())

# Tambahkan hasil sentimen ke kolom 'category'
df['category'] = [result['label'] for result in results]

# Simpan dataset yang sudah diberi label ke file CSV baru
output_file_path = "IKD_Dataset_Processed_Labelled.csv"
df.to_csv(output_file_path, index=False)

print(f"Dataset berhasil diberi label dan disimpan ke {output_file_path}")
```
